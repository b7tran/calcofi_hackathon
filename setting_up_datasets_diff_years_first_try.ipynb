```python
xlsx=pd.ExcelFile("definitions.xlsx")
```

```python
data=pd.read_excel(xlsx,"Species Codes")
data
```

```python
codes=data["Code"].tolist()
```

```python
names=data["Species"].tolist()
```

```python
species_codes= [x for x in codes if str(x) != 'nan']
species_names= [x for x in names if str(x) != 'nan']
```

```python
len(species_codes)==len(species_names)
```

```python
lst=[]
for number in species_codes:
    lst.append(int(number))
```

```python
lst
```

```python
dictionary=dict(zip(lst,species_names))
```

```python
dictionary
```
